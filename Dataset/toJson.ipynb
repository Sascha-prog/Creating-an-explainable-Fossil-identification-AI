```python
import pandas as pd
import json

# ---- CONFIG ----
excel_file = "csv as a nice table.xlsx"     # your Excel file
json_file = "dataset.json"     # your output JSON file

# Read Excel
df = pd.read_excel(excel_file)

# Expect columns: filename, description
# Adjust names here if your column headers differ
df.columns = [col.lower() for col in df.columns]  # normalize headers

key_col = "filename"
value_col = "reviewer_notes"

# Build dictionary
output_dict = {
    str(row[key_col]): str(row[value_col])
    for _, row in df.iterrows()
}

# Write JSON
with open(json_file, "w", encoding="utf-8") as f:
    json.dump(output_dict, f, indent=4, ensure_ascii=False)

print("JSON written to", json_file)
```
